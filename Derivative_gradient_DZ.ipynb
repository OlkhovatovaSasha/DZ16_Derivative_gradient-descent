### <center> Задание 4:
Градиентный спуск своими руками

```python
def f(x):
    return (x[0]**2) + (x[1]**2)
```

```python
x = np.array([100, 200])
eps = np.sqrt(np.finfo(float).eps)
lr = 0.1
```

```python
for i in range(100):
    # Calculate gradient
    grad = approx_fprime(x, f, [eps, eps]) # Your code here (функция для вычисления градиента есть в лекционном ноутбуке)
    # Update x with gradient
    x = x - lr*(grad*(-1))  # Your code here (смотри слайд презентации про градиентный спуск)
```

```python
print("Минимум достигается в: ", x)
print("Значение функции в минимуме: ", f(x))
# Проверка с помощью встроенной функции scipy: не понимаю как это делать
```
